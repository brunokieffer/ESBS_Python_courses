## Drawing with a turtle

The purpose of thess exercise is to draw complex shapes from simple elementary drawing functions (such as drawing a square). We will use a library call turtle, which allows the drawing of simple 2D graphics using a basic set of graphics commands which are described [here](https://docs.python.org/2/library/turtle.html). The commands defines how a pen is moving, by specifying the length of path, the direction (heading) of motion and if the path is traced (pendown()) or not (penup()). Alternatively, we can tell the turtle to go to a specific position by providing its 2D coordinates as a tuple (x,y).

The most useful commands are :   
  
* forward(a_length) ===> move forward for a distance a_length
  
* backward(a_length) ===>  move backward for a distance a_length
  
* rigth(my_angle) or left(my_angle) ===>  turn the heading to the right (or left) by an angle my_angle
  
* circle(a_length) ===>  draw a circle of radius a_length at current position
  
* color(color_string) ===>  change the color of the ink (color_string = 'blue' or 'red' etc ....)
